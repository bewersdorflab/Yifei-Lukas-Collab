```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import os
import csv

#Color blind friendly colors to use in plots
CB_color_cycle = ['#377eb8', '#ff7f00', '#4daf4a',
                  '#f781bf', '#a65628', '#984ea3',
                  '#999999', '#e41a1c', '#dede00']

#Read .csv file and convert to numpy array
read_csv = pd.read_csv(data)
read_csv = np.array(read_csv)
```

```python
reps = int(np.size(read_csv,1)/2) #Calculate number of unique line plots

#Extract distance colum and grey values column and remove NaNs from both
fig = plt.figure()
for a in range(reps):
    dist = read_csv[:,0+(2*a)]
    gv = read_csv[:,1+(2*a)]
    dist = dist[~np.isnan(dist)]
    gv = gv[~np.isnan(gv)]
    #Normalize the grey values
    gv_max = np.max(gv)
    gv_norm = gv/gv_max
    plt.plot(dist, gv_norm, color=CB_color_cycle[a])
plt.show()
```

```python
dist = read_csv[:,2]
gv = read_csv[:,3]
dist = dist[~np.isnan(dist)]
gv = gv[~np.isnan(gv)]
gv_max = np.max(gv)
gv_norm = gv/gv_max
```

```python
plt.plot(dist,gv_norm,color=CB_color_cycle[0])
plt.show()
```
